```python
# 1. Install vLLM
!pip install vllm --upgrade

# 2. Pick a very small model
# We'll use "TinyLlama/TinyLlama-1.1B-Chat-v1.0" (1.1 billion params)

# 3. Launch vLLM server
from vllm import LLM
from vllm import SamplingParams

# Initialize the LLM
llm = LLM(
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    dtype="float16",  # Use float16 to fit easily on small GPUs
    trust_remote_code=True,  # Allow loading custom code if needed
)

# Create a sampling parameters object
sampling_params = SamplingParams(
    temperature=0.7,
    top_p=0.9,
    max_tokens=128,
)

# 4. Make a simple demo query
prompts = [
    "What is the capital of France?",
    "Explain quantum physics in simple words.",
]

# Generate completions
outputs = llm.generate(prompts, sampling_params)

# 5. Print results
for output in outputs:
    print(f"Prompt: {output.prompt}")
    print(f"Generated: {output.outputs[0].text.strip()}")
    print("="*50)
```
